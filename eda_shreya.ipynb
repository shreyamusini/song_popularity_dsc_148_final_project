```python
def most_common_genre(genre_string):
    genres = genre_string.split(", ") 
    return max(set(genres), key=genres.count)
```

### Track Genre Proxy

```python
df_grouped = df_total.groupby(['track_artist', 'track_name'])['playlist_genre'].agg(", ".join)
df_grouped = df_grouped.apply(most_common_genre)
df_total['track_genre_pseudo'] = df_total.set_index(['track_artist', 'track_name']).index.map(df_grouped)
df_total


```

```python
track_genre_count = df_total.groupby('track_genre_pseudo')['track_genre_pseudo'].count()
fig5 = px.bar(track_genre_count, labels = {'index': 'Track Genre', 'value': 'Count'},
              title = 'Track Genre Count')
fig5
```

```python
## Popularity Classification (High: > 68, Low: <= 68)
df_total['Popularity Classification'] = df_total['track_popularity'].apply(lambda x: 'high' if x > 68 else 'low')
df_total
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
category_encoder = LabelEncoder()
category_encoder.fit(df_total['playlist_subgenre'])
```

```python
track_genre_encoder = LabelEncoder()
track_genre_encoder.fit(df_total['track_genre_pseudo'])
```

```python
vectorizer_artist = TfidfVectorizer()
X_artist_names = vectorizer_artist.fit_transform(df_total['track_artist'])
```

```python
df_cop = df_total.drop(['Popularity Classification', 'playlist_name', 'playlist_genre', 'id', 'uri', 'analysis_url', 'track_href', 'type', 'track_album_id', 'playlist_id', 'track_id', 'track_name', 'track_album_name', 'track_popularity'], axis = 1)
y = df_total['Popularity Classification']
df_cop['playlist_subgenre'] = category_encoder.transform(df_cop['playlist_subgenre'])
df_cop['track_genre_pseudo'] = track_genre_encoder.transform(df_cop['track_genre_pseudo'])
df_cop['track_album_release_date'] = pd.to_datetime(df_cop['track_album_release_date'], errors='coerce').dt.year.fillna(0)
```

```python
numerical_cols = df_cop.select_dtypes(include=['number'])
df_cop[numerical_cols.columns] = numerical_cols.fillna(numerical_cols.median())
# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(X)
X = df_cop[numerical_cols.columns].to_numpy()
X
```

```python
from scipy.sparse import csr_matrix, hstack
X_sparse = csr_matrix(X)
```

```python
X_combined = hstack([X_artist_names, X])
X_combined
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

scaler = StandardScaler(with_mean = False)
train_x_scaled = scaler.fit_transform(X_train)  
test_x_scaled = scaler.transform(X_test)

model = LogisticRegression(max_iter=500)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
